```python
import sqlite3
from pathlib import Path
import numpy as np
import pandas as pd
```

```python
db_path = Path("../../data/database.sqlite")

conn = sqlite3.connect(db_path.as_posix())

df = pd.read_sql_query("SELECT * FROM Match_Aggregated", conn)

conn.close()
```

# Extract Player Stats

```python
player_stats_cols = [
    "preferred_foot",
    "attacking_work_rate",
    "defensive_work_rate",
    "crossing",
    "finishing",
    "heading_accuracy",
    "short_passing",
    "volleys",
    "dribbling",
    "curve",
    "free_kick_accuracy",
    "long_passing",
    "ball_control",
    "acceleration",
    "sprint_speed",
    "agility",
    "reactions",
    "balance",
    "shot_power",
    "jumping",
    "stamina",
    "strength",
    "long_shots",
    "aggression",
    "interceptions",
    "positioning",
    "vision",
    "penalties",
    "marking",
    "standing_tackle",
    "sliding_tackle",
]
```

```python
pd.set_option("display.max_columns", None)
df[player_stats_cols].describe()
```

```python
# Scale from 0 - 2 to -1 - 1
df[["attacking_work_rate", "defensive_work_rate"]] = (
    df[["attacking_work_rate", "defensive_work_rate"]] - 1
)

# Scale from 0 - 100 to -1 - 1
scale_cols = [
    "crossing",
    "finishing",
    "heading_accuracy",
    "short_passing",
    "volleys",
    "dribbling",
    "curve",
    "free_kick_accuracy",
    "long_passing",
    "ball_control",
    "acceleration",
    "sprint_speed",
    "agility",
    "reactions",
    "balance",
    "shot_power",
    "jumping",
    "stamina",
    "strength",
    "long_shots",
    "aggression",
    "interceptions",
    "positioning",
    "vision",
    "penalties",
    "marking",
    "standing_tackle",
    "sliding_tackle",
]
df[scale_cols] = (df[scale_cols] / 50) - 1

df_player = df[player_stats_cols]
```

```python
df_player.describe()
```

# Extract Formation Stats

```python
formation_data = []
for (_, team_side), df_group in df.groupby(["match_api_id", "team_side"]):
    # Extract relevant Columns
    df_player = df_group[player_stats_cols]
    df_position = df_group[["player_x", "player_y"]]

    # Normalize Player Positions
    np_position = df_position.to_numpy()
    np_position_normalized = (np_position - np_position.min(axis=0)) / (
        np_position.max(axis=0) - np_position.min(axis=0)
    )

    for i in range(np_position.shape[0]):
        # Extract Player
        current_player = np_position_normalized[i]
        other_players = np.delete(np_position_normalized, i, axis=0)

        # Calculate Distance
        relative_vectors = other_players - current_player
        distances = np.linalg.norm(relative_vectors, ord=2, axis=1)

        # Calculate Angle
        angles_rad = np.arctan2(relative_vectors[:, 1], relative_vectors[:, 0])

        formation_data.append(np.hstack([current_player, distances, angles_rad]))
```

```python
formation_cols = (
    ["x", "y"]
    + [f"distance_{i + 1}" for i in range(9)]
    + [f"angle_{i + 1}" for i in range(9)]
)
df_formation = pd.DataFrame(formation_data, columns=formation_cols)
```

```python
df_formation.head()
```

# Concat Dataframes

```python
df_final = pd.concat([df_player, df_formation], axis=1)
```

```python
df_final.head()
```

```python
conn = sqlite3.connect(db_path.as_posix())

df_final.to_sql(
    "Player_Formation_Preprocessed",
    conn,
    if_exists="replace",
    index=False,
)

conn.close()
```
